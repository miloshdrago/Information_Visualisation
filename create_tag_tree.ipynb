```python
def add_label(d, l):
    val = l.pop(0)
    try:
        index = list(map(lambda x: x['name'], d)).index(val)
    except:
        d.append({'name': val, 'children':[], 'values':[]})
        index = len(d) - 1
    if len(l):
        add_label(d[index]['children'], l)
```

```python
tree = {'name':'root', 'children':[], 'values':[]}
for k,v in sub_tags.items():
    add_label(tree['children'], v)
```

# Load Data

```python
with open('data/tags/sub_tags.json', 'r') as f:
    sub_tags = json.load(f)
```

```python
# Adjust some catagories

# Merge Architecture into Other
l = [x for x in sub_tags.items() if x[1][0] == 'Architecture']
for k,v in l:
    sub_tags[k][0] = 'Other'
```

# Tree Algo

```python
def add_label(d, l, s):
    val = l.pop(0)
    try:
        index = list(map(lambda x: x['name'], d['children'])).index(val)
    except:
        d['children'].append({'name': val, 'children':[], 'values':[]})
        index = len(d['children']) - 1
    if len(l):
        add_label(d['children'][index], l, s)
    else:
        d['children'][index]['values'].append(s)
```

```python
tree = {'name':'root', 'children':[], 'values':[]}
for k,v in sub_tags.items():
    add_label(tree, v, k)
tree
```

```python
with open('data/tags/tag_hierarchical.json', 'w') as f:
    json.dump(tree, f)
```

# Filter File

```python
reverse_tags = {}
for k,v in sub_tags.items():
    for i in v:
        reverse_tags[i] = set()
```

```python
for k,v in sub_tags.items():
    for i in v:
        reverse_tags[i].add(k)
reverse_tags
```

```python
with open('data/tag_filter.json', 'w') as f:
    dump_dict = {}
    for k,v in reverse_tags.items():
        dump_dict[k] = list(v)
    json.dump(dump_dict, f)
```
